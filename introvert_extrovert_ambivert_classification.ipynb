```python
# Categorize based on score (adjust thresholds as needed)
def categorize(score):
    if score <= 3:
        return 'Introvert'
    elif score <= 7:
        return 'Ambivert'
    else:
        return 'Extrovert'

empathy_df['emotional_type'] = empathy_df['emotional_stability'].apply(categorize)

# Count categories
counts = empathy_df['emotional_type'].value_counts()[['Introvert', 'Ambivert', 'Extrovert']]

# Plot
plt.figure(figsize=(8,4))
plt.bar(counts.index, counts.values, color=['#dfdfdf', 'blue', 'grey'])
plt.xlabel('Emotional Type')
plt.ylabel('Count')
plt.title('Categorized Emotional Stability')
plt.show()
```

```python
empathy_df.rename(columns={"creativity\t": "creativity"}, inplace=True)
```

```python
# Bersihkan whitespace dari nama kolom
empathy_df.columns = empathy_df.columns.str.strip()
```

```python
# Lakukan agregasi
agg_df = empathy_df.groupby('personality_type').agg({
    'spontaneity': 'nunique',
    'creativity': ['min', 'max']
})

# Rename kolom untuk kejelasan
agg_df.columns = ['unique_spontaneity', 'min_creativity', 'max_creativity']
agg_df = agg_df.reset_index()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(len(agg_df['personality_type']))  # posisi X
width = 0.25  # lebar bar

plt.figure(figsize=(10, 6))

# Plot masing-masing metrik
plt.bar(x - width, agg_df['unique_spontaneity'], width, label='Spontaneity (unique)', color='skyblue')
plt.bar(x, agg_df['min_creativity'], width, label='Creativity (min)', color='lightgreen')
plt.bar(x + width, agg_df['max_creativity'], width, label='Creativity (max)', color='orange')

# Label & tampilan
plt.xlabel('Personality Type')
plt.ylabel('Value')
plt.title('Summary by Personality Type')
plt.xticks(x, agg_df['personality_type'], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
